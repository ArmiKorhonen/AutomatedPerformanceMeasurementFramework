```python
# Use Pandas for data processing and Matplotlib for visualization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv file into dataframe
df = pd.read_csv('vr_data_2023.csv')
```

```python
# Show the amount of headsets in dataset and the first five rows
print(df.shape[0])
df.head()
```

```python
# Store info on Meta Quest 3 and Meta Quest 2 from the dataset and show them
quest3 = df[df['headset'] == 'Meta Quest 3']
quest2 = df[df['headset'] == 'Oculus Quest 2']
print(quest3)
print(quest2)
```

```python
# Create a scatter plot with colormap
scatter = plt.scatter(df['refresh_rate'], df['total_pixels'], c=df['year'], cmap='viridis')

# Adding a color bar
plt.colorbar(scatter, label='year')

# Adding title and labels
plt.title('Refresh rate vs Total Pixels (Colored by Year)')
plt.xlabel('Refresh rate (Hz)')
plt.ylabel('Total Pixels')

# Plot the previously stored points for Meta Quest 3 and Meta Quest 2 with a distinct edgecolor
plt.scatter(quest3['refresh_rate'], quest3['total_pixels'], 
            edgecolor='red', facecolors='none', label='Meta Quest 3', zorder=5)
plt.scatter(quest2['refresh_rate'], quest2['total_pixels'], 
            edgecolor='blue', facecolors='none', label='Meta Quest 2', zorder=5)

# Show the legend with a smaller font size in the lower right corner
plt.legend(fontsize='small', loc='lower right')

# Save the plot as a pdf-file
plt.savefig('vr_data_2023_scatter_plot.pdf', bbox_inches='tight')

# Show the plot
plt.show()
```
