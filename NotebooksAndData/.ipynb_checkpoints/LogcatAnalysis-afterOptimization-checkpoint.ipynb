```python
import pandas as pd

# Step 1: Read the log file
log_path = '15-05-2024/logcat-03b.txt'  # Update this path as needed
with open(log_path, 'r') as file:
    log_lines = file.readlines()

# Helper function to extract the section name including test run details
def extract_section_name(line):
    # Split the line and extract relevant parts
    parts = line.strip().split()
    if "Starting test run" in line:
        # Example log: "05-08 10:43:24.205 30649 30670 I Unity : Starting test run 3 for VFX with passthrough False"
        run_index = parts.index("run") + 1
        run_number = parts[run_index]  # Run number
        for_index = parts.index("for") + 1
        with_index = parts.index("with")
        test_description = ' '.join(parts[for_index:with_index])  # Captures 'VFX', 'BuiltIn', or 'none'
        passthrough_status = parts[-1]  # Captures 'True' or 'False'
        return f"{test_description} with passthrough {passthrough_status} run {run_number}"
    return "Undefined"


# Step 2: Parse the log to extract sections
sections = []
current_section = []
section_names = []

for line in log_lines:
    if 'Starting test run' in line:
        if current_section:  # Close current section before starting new
            sections.append(current_section)
            current_section = []
        section_name = extract_section_name(line)
        section_names.append(section_name)
        current_section.append(line)
    elif 'Ended test run' in line:
        current_section.append(line)
        sections.append(current_section)  # Save the complete section
        current_section = []  # Reset for the next section
    elif current_section:
        current_section.append(line)

if current_section:
    sections.append(current_section)  # Append the last section if any

# Step 3: Save each section to a separate file and process into DataFrames
dataframes = {}

def parse_log_line(log_line):
    parts = log_line.split(' ')
    timestamp = parts[0] + ' ' + parts[1]
    message_type = parts[4]
    data = parts[5:]
    data_str = ' '.join(data)
    data_pairs = data_str.split(',')
    data_dict = {pair.split('=')[0]: pair.split('=')[1] for pair in data_pairs if '=' in pair}
    data_dict['Timestamp'] = timestamp
    data_dict['MessageType'] = message_type
    return data_dict

def process_log_section(lines):
    data_rows = []
    for line in lines:
        if 'VrApi' in line:  # Filter for specific log messages
            parsed_data = parse_log_line(line)
            data_rows.append(parsed_data)
    return pd.DataFrame(data_rows)

for index, (section, name) in enumerate(zip(sections, section_names)):
    section_path = f'Data/{name.replace(":", "").replace(" ", "_").replace("/", "-")}.log'
    with open(section_path, 'w') as file:
        file.writelines(section)
    
    df = process_log_section(section)
    dataframes[name] = df

# Print summary of DataFrames created
for name, df in dataframes.items():
    print(f"DataFrame for {name} created with {len(df)} records.")
    print(df.head())  # Optionally print first few rows of each DataFrame

```

```python
# Selected columns that you want in all DataFrames
selected_columns = ['Timestamp', 'FPS', 'Stale', 'CPU4/GPU', 'App', 'Mem', 'Free', 'CPU&GPU', 'GPU%', 'CPU%']

# Dictionary to store the cleaned and filtered DataFrames
cleaned_dataframes = {}

for name, df in dataframes.items():
    # Strip whitespace from all column names
    df.columns = [col.strip() for col in df.columns]
    
    # Rename specific columns
    # Example assumes that there's a consistent naming issue to fix; adjust as necessary
    df.rename(columns={'VrApi   : FPS': 'FPS'}, inplace=True)
    
    # Filter the DataFrame to include only the selected columns, checking if they exist
    filtered_columns = [col for col in selected_columns if col in df.columns]
    df_filtered = df[filtered_columns]

    # Store the cleaned, filtered DataFrame back in a new dictionary
    cleaned_dataframes[name] = df_filtered

    # Optionally, print out the first row to verify the DataFrame
    print(f"Processed DataFrame for {name}:")
    print(df_filtered.head(1))
    print()  # Print a newline for better separation in output

# If you want to overwrite the original dataframes with the cleaned ones, you can uncomment the next line:
# dataframes = cleaned_dataframes

```

```python
import re

# Define function to extract numerical part from string
def extract_number(input_string):
    number = re.findall(r'\d+', input_string)
    return int(number[0]) if number else None

def extract_primary_cpu_percentage(input_string):
    number = re.search(r'^\d+\.?\d*', input_string)  # Allow decimal numbers
    return float(number.group()) if number else None



# Process each DataFrame in the cleaned dataframes dictionary
for name, df in cleaned_dataframes.items():
    print(f"Columns before transformation for {name}: {df.columns.tolist()}")
    # Apply extract_number to specific columns
    df['App'] = df['App'].apply(extract_number)
    df['Mem'] = df['Mem'].apply(extract_number)
    df['Stale'] = df['Stale'].apply(extract_number)
    df['Free'] = df['Free'].apply(extract_number)
    df['CPU&GPU'] = df['CPU&GPU'].apply(extract_number)
    
    # Convert GPU% and CPU% columns to float where applicable
    df['GPU%'] = df['GPU%'].apply(lambda x: float(x))
    df['CPU%'] = df['CPU%'].apply(extract_primary_cpu_percentage)

    # Assuming 'FPS' column exists and contains data in the format "current/max"
    if 'FPS' in df.columns:
        df['Current FPS'] = df['FPS'].str.split('/').str[0].astype(int)

    # Rename columns to include units where appropriate
    df.rename(columns={
        'App': 'App (ms)',
        'Mem': 'Mem (MHz)',
        'Free': 'Free (MB)',
        'CPU&GPU': 'CPU&GPU (ms)',
        'GPU%': 'GPU (%)',
        'CPU%': 'CPU (%)'
    }, inplace=True)
    # Convert the 'Timestamp' column from string to datetime objects
    if 'Timestamp' in df.columns:
        df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%m-%d %H:%M:%S.%f', errors='coerce')

        # Convert 'Timestamp' from datetime to seconds from the start of the recording
        df['Seconds From Start'] = (df['Timestamp'] - df['Timestamp'].min()).dt.total_seconds()
     
    # Optionally print out the first few rows to verify the DataFrame
    print(f"Processed DataFrame for {name}:")
    print(df.head())
    print()  # Print a newline for better separation in output

    # Optionally print out the first few rows to verify the DataFrame
    print(f"Processed DataFrame for {name} with timestamps converted to seconds from start:")
    print(df[['Timestamp', 'Seconds From Start']].head())
    print()  # Print a newline for better separation in output
```

```python
import pandas as pd

# Step 1: Organize DataFrames into groups based on test configuration
grouped_dataframes = {}
for name, df in cleaned_dataframes.items():
    # This assumes the name format is 'TestType with passthrough Status run Number'
    config_name = ' '.join(name.split()[:-2])  # Remove 'run Number' to group by test configuration
    if config_name not in grouped_dataframes:
        grouped_dataframes[config_name] = []
    grouped_dataframes[config_name].append(df)

# Function to check if a column can be converted to numeric
def can_be_numeric(series):
    try:
        pd.to_numeric(series)
        return True
    except ValueError:
        return False

# Step 2: Calculate the average DataFrame for each configuration
averaged_dataframes = {}
for config, dfs in grouped_dataframes.items():
    # Concatenate DataFrames along axis=0 (stack them vertically)
    concatenated_df = pd.concat(dfs)
    
    # Filter out non-numeric columns
    numeric_df = concatenated_df.loc[:, concatenated_df.apply(can_be_numeric)]
    
    # Convert columns to numeric
    numeric_df = numeric_df.apply(pd.to_numeric, errors='coerce')
    
    # Group by the index and calculate the mean
    averaged_df = numeric_df.groupby(numeric_df.index).mean()
    averaged_dataframes[config] = averaged_df

# Now you have a dictionary 'averaged_dataframes' where each key is a configuration and each value is the averaged DataFrame

# Optionally print out the first few rows to verify the averaged DataFrame
for config, averaged_df in averaged_dataframes.items():
    print(f"Averaged DataFrame for {config}:")
    print(averaged_df.head())
    print()  # Print a newline for better separation in output
```

```python
for name, df in averaged_dataframes.items():
    print(f"Description of {name}:")
    print(df.describe())
    print("\n")  # Adds a newline for better separation
```

```python
descriptive_titles = {
    'VFX with passthrough False': "VFX particle effect, immersive mode",
    'VFX with passthrough True': "VFX particle effect, passthrough mode",
    'BuiltIn with passthrough False': "Built-in particle effect, immersive mode",
    'BuiltIn with passthrough True': "Built-in particle effect, passthrough mode",
    'none with passthrough False': "No particle effect, immersive mode",
    'none with passthrough True': "No particle effect, passthrough mode",
}
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for name, df in averaged_dataframes.items():
    if 'GPU (%)' in df.columns and 'Seconds From Start' in df.columns:
        description = descriptive_titles.get(name, "Unknown Configuration")  # Default to "Unknown Configuration" if key not found

        # Normalize time to percentage of total duration
        total_duration = df['Seconds From Start'].max()
        df['Time Normalized'] = (df['Seconds From Start'] / total_duration) * 100

        plt.plot(df['Time Normalized'], df['GPU (%)'], label=f'{description}', marker='o', linestyle='-', markersize=4)

plt.title('GPU Usage Over Normalized Time Across All Sections')
plt.xlabel('Normalized Time (%)')
plt.ylabel('GPU Usage (%)')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
